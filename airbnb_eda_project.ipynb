### EDA

**5. Analise de dados**

**Analise univariada**

```python
#identificando outliers 

sns.boxplot(data= data, x= 'price')
```

```python
df = data[data['price'] < 1500]

sns.boxplot(data= df, x= 'price')
```

```python
#distribuição da variavel Price

plt.figure(figsize = (8, 5))
sns.histplot(data= df, x= 'price', bins= 100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
data.columns
```

```python
# Distribuição da variavel availability_365 

plt.figure(figsize = (6, 3))
sns.histplot(data= df, x= 'availability_365', bins= 100)
plt.title("'availability_365' Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
#preço médio por bairo

df.groupby(by = 'neighbourhood_group')['price'].mean().sort_values(ascending= True)
```

**Feature Engineering**

```python
#criando uma nova coluna: preço por quantidade de camas

df['price per bed'] = df['price']/df['beds']

df.head()
```

```python
#preço médio por bairro

df.groupby(by = 'neighbourhood_group')['price per bed'].mean().sort_values(ascending= True)
```

**Analise Bivariada** Dependência de uma variável em outra variável.

```python
df.columns
```

```python
#dependencia da variavel price sobre neighbourhood_group

sns.barplot(data=df, x ='neighbourhood_group', y = 'price', hue= 'room_type' )
```

```python
# relação do numero de avaliações com o preço

plt.figure(figsize= (8, 5))
plt.title("Dependência entre Localização e Avaliações")
sns.scatterplot(data = df, x = 'number_of_reviews', y= 'price', hue = 'neighbourhood_group')
plt.show()
```

```python
#Análise de Correlação entre Variáveis por Tipo de Quarto
sns.pairplot(data= df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365' ], hue = 'room_type')
```

```python
#Distribuição geográfica dos AirBnB Listing

plt.figure(figsize=(10,7))
plt.title("Distribuição Geográfica das Listagens do Airbnb")
sns.scatterplot(data = df, x = 'longitude', y= 'latitude', hue = 'room_type') 
plt.show()
```

```python
df.dtypes
```

```python
#Mapa de calor - correlação entre as variáveis numéricas

corr = df[['latitude', 'longitude', 'beds', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365' ]].corr()

plt.figure(figsize = (8, 6))
sns.heatmap(data = corr, annot = True)
```
